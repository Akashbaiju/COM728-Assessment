```python
import csv
import sys

def query_a2(records, absence_threshold, parental_involvement_threshold):
    result = []
    for row in records:
        if int(row[6]) < absence_threshold and int(row[9]) >= parental_involvement_threshold:
            student_info = {
                'ID': row[0],
                'Free_time': int(row[7]),
                'Math_score': int(row[3]),
                'Reading_score': int(row[4]),
                'Writing_score': int(row[5])
            }
            result.append(student_info)
    return result

# Get user input for parental involvement threshold
parental_involvement_threshold = int(input("Enter the parental involvement threshold: "))

# Load the student records
records = []
with open('student_data.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        records.append(row)

# Call the query_a2 function with the user-provided threshold
result = query_a2(records, 50, parental_involvement_threshold)

# Print the results
for student in result:
    print(student)
```

```python
import csv

# Open the CSV file
with open('students_data.csv', 'r') as file:
    reader = csv.DictReader(file)
    
    # Initialize an empty list to store the student data
    student_data = []
    
    # Iterate through the rows in the CSV file
    for row in reader:
        # Check if the student has less than 50 absences
        if int(row['Absences']) < 50:
            # Extract the requested information
            student_id = row['ID']
            freetime = int(row['Freetime'])
            math_score = int(row['Math_score'])
            reading_score = int(row['Reading_score'])
            writing_score = int(row['Writing_score'])
            parental_involvement = row['Parental_involvement']
            
            # Add the student data to the list
            student_data.append({
                'ID': student_id,
                'Freetime': freetime,
                'Math_score': math_score,
                'Reading_score': reading_score,
                'Writing_score': writing_score,
                'Parental_involvement': parental_involvement
            })
    
    # Print the student data based on parental involvement
    print("Students with less than 50 absences:")
    for student in student_data:
        if student['Parental_involvement'] == 'low':
            print(f"ID: {student['ID']}, Freetime: {student['Freetime']}, Math_score: {student['Math_score']}, Reading_score: {student['Reading_score']}, Writing_score: {student['Writing_score']}")
        elif student['Parental_involvement'] == 'medium':
            print(f"ID: {student['ID']}, Freetime: {student['Freetime']}, Math_score: {student['Math_score']}, Reading_score: {student['Reading_score']}, Writing_score: {student['Writing_score']}")
        elif student['Parental_involvement'] == 'high':
            print(f"ID: {student['ID']}, Freetime: {student['Freetime']}, Math_score: {student['Math_score']}, Reading_score: {student['Reading_score']}, Writing_score: {student['Writing_score']}")
```

```python
import csv

# Open the CSV file
with open('students_data.csv', 'r') as file:
    reader = csv.DictReader(file)
    
    # Get the user input for student age
    student_age = int(input("Enter the student's age: "))
    
    # Initialize an empty list to store the student data
    student_data = []
    
    # Iterate through the rows in the CSV file
    for row in reader:
        # Check if the student's age matches the user input
        if int(row['Age']) == student_age:
            # Extract the requested information
            student_id = row['ID']
            family_size = int(row['Family_size'])
            traveltime = int(row['Traveltime'])
            health = row['Health']
            
            # Add the student data to the list
            student_data.append({
                'ID': student_id,
                'Family_size': family_size,
                'Traveltime': traveltime,
                'Health': health
            })
    
    # Print the student data for students with the specified age
    print(f"Students with age {student_age}:")
    for student in student_data:
        print(f"ID: {student['ID']}, Family_size: {student['Family_size']}, Traveltime: {student['Traveltime']}, Health: {student['Health']}")
```
